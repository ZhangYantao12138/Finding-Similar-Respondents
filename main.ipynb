## 定义一些相关函数

```python
# 定义一个函数，找出目标姓名所在的行
def find_target_data(dataframe, name):
    target_row = dataframe[dataframe['1、你的姓名'] == name]
    if len(target_row) > 0:
        return target_row.iloc[0]
    else:
        print('未找到该姓名对应的数据')
        return None
```

```python
# 距离法（欧氏距离）
def dis(df,target_row):
    # 删去序号、姓名、学号和提示信息，建立一个新df_new
    df_new = df.iloc[:, 4:]
    target_row_processed = target_row.iloc[4:]
    # 计算欧几里得距离
    distances = np.sqrt(np.sum((df_new.iloc[:, :] - target_row_processed) ** 2, axis=1))
    # 找出最相关的一条
    most_similar_index_1 = distances.idxmin()
    most_similar_row_1 = df.iloc[most_similar_index_1, 1]
    # 返回姓名和相关度
    return most_similar_row_1,distances[most_similar_index_1]
```

```python
# 皮尔逊系数法
def p(df,target_row):
    # 删去序号、姓名、学号和提示信息，建立一个新df_new
    df_new = df.iloc[:, 4:]
    # 统一目标行维度
    target_row_processed = target_row.iloc[4:]
    # 处理NaN值
    df_new = df_new.fillna(0)
    target_row_processed = target_row_processed.fillna(0)
    # 计算皮尔逊相关系数
    pearson_correlations = df_new.corrwith(target_row_processed, axis=1)
    pearson_correlations = pearson_correlations.dropna()  # 删除含有NaN的相关系数
    most_similar_index_2 = pearson_correlations.idxmax()
    most_similar_row_2 = df.iloc[int(most_similar_index_2), 1]
    # 返回姓名和相关度
    return most_similar_row_2, pearson_correlations[most_similar_index_2]
```

```python
# 余弦法
def cos(df,target_row):
    # 删去序号、姓名、学号和提示信息，建立一个新df_new
    df_new = df.iloc[:, 4:]
    # 统一目标行维度
    target_row_processed = target_row.iloc[4:]
    # 计算余弦相似度
    cosine_similarities = df_new.apply(lambda row: np.dot(row, target_row_processed) / (np.linalg.norm(row) * np.linalg.norm(target_row_processed)), axis=1)
    # 找出最相关的一条
    most_similar_index_3 = cosine_similarities.idxmax()
    most_similar_row_3 = df.iloc[most_similar_index_3, 1]
    # 返回姓名和相关度
    return most_similar_row_3, cosine_similarities[most_similar_index_3]
```

## 主程序

```python
# 主程序
import pandas as pd
import numpy  as np
# 读取.xlsx文件
data = pd.read_excel('推荐系统调查表.xlsx')
# 获取姓名信息
your_name = input('请输入姓名')
# your_name = '章言韬'
# 选择要计算相关性的数据行
target_row = find_target_data(data,your_name)
# 在数据表中删去目标行
df = data.drop(target_row.name)
# 分别用三种方法计算，获取最相关的同学名字以及相关性
dis_output1, dis_output2 = dis(df,target_row)
p_output1, p_output2 = p(df,target_row)
cos_output1, cos_output2 = cos(df,target_row)
# 输出结果
print(f"{your_name}同学:\n根据欧式距离法:与你爱好最相似的同学是{dis_output1}，欧氏距离为{dis_output2}；\n根据皮埃尔相关系数法:与你爱好最相似的同学是{p_output1}，相关系数为{p_output2};\n根据余弦法:与你爱好最相似的同学是{cos_output1}，余弦相似度为{cos_output2}；\n")
```
